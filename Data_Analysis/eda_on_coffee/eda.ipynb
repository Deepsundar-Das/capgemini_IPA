```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Coffee.csv")
```

Which columns contain missing values?

```python
df.isnull().sum()
df.columns[df.isnull().any()]
```

How many unique countries are present in the dataset?

```python
df["Country"].nunique()
```

How many records are available for each year?

```python
df["Year"].value_counts().sort_index()
```

How many records are available for each country?

```python
df["Country"].value_counts()
```

Find the minimum, maximum, and average of total production.

```python
df["Total Production"].agg(["min", "max", "mean"])
```

Which country has the highest total production in a single record?

```python
df.loc[df["Total Production"].idxmax()]
```

Are there any rows where total production is missing or zero?

```python
df[(df["Total Production"].isnull()) | (df["Total Production"] == 0)]
```

What is the average import price across all records?

```python
df["Import Price"].mean()
```

What is the total coffee production for each year?

```python
df.groupby("Year")["Total Production"].sum()
```

Which are the top 5 countries based on total production?

```python
df.groupby("Country")["Total Production"].sum().sort_values(ascending=False).head(5)
```

Which country has the highest average export price?

```python
df.groupby("Country")["Export Price"].mean().sort_values(ascending=False).head(1)
```

Does domestic consumption exceed total production for any country?

```python
len(df[df["Domestic Consumption"] > df["Total Production"]]) > 0
```

Which country shows the highest growth in total production over the years?

```python
df_sorted = df.sort_values("Year")
growth = df_sorted.groupby("Country")["Total Production"].agg(lambda x: x.iloc[-1] - x.iloc[0])
growth.sort_values(ascending=False)
```

Is there any relationship between export price and total production?

```python
df["Export Price"].corr(df["Total Production"])
```
