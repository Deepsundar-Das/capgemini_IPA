#### **Merge and Concatinate**

```python
data1 = {
    'Name':['A','B','C'],
    'age':[21,45,32],
    'salary':[10000, 200000, 60000000]
}

data2 = {
     'Name':['A1','B1','C'],
    'age':[56,35,45],
    'salary':[75288, 455555, 987668865]
}

new_df1 = pd.DataFrame(data1)
new_df2 = pd.DataFrame(data2)
```

```python
new_df1.head()
```

```python
new_df2.head()
```

```python
pd.concat([new_df1, new_df2], ignore_index=True)
```

```python
pd.concat([new_df1, new_df2], ignore_index=True, axis = 1)
```

```python
new_df = pd.concat([new_df1, new_df2], ignore_index=True)
```

```python
new_df.drop_duplicates(inplace=True)
new_df.head()
```

#### **Merge**

```python
pd.merge(new_df1, new_df2, on='salary', how='outer')
```
