#### **Matplotlib**

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
x = np.random.randint(1,50,10)
y = np.random.randint(1, 50, 10)
print(x, y, sep='\n')
```

```python
plt.plot(x, y, color = 'brown', linestyle='dotted')
plt.show()
```

```python
plt.title('Random chart')
plt.plot(x, y, color = 'brown', linestyle='dotted')
plt.show()
```

```python
plt.title('Random chart')
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.plot(x, y, color = 'brown', linestyle='dotted')
plt.show()
```

```python
plt.title('Random chart')
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'brown', linestyle='dotted')
plt.show()
```

```python
plt.title('Random chart')
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'orange', linestyle=':')
plt.show()
```

```python
plt.title('Random chart')
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'orange', linestyle=':', alpha=0.9)
plt.show()
```

```python
plt.title('Random chart')
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'orange', linestyle=':', alpha=0.9, marker="*")
plt.show()
```

```python
plt.title('Random chart')
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'orange', linestyle=':', alpha=0.9, marker="*", markerfacecolor='y')
plt.show()
```

```python
plt.title('Random chart')
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'orange', linestyle=':', alpha=0.9, marker="*", markerfacecolor='y', markersize=20)
plt.show()
```

```python
plt.title('Random chart')
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'orange', linestyle=':', alpha=0.9, marker="*", markerfacecolor='y', markersize=20, markeredgecolor='blue')
plt.show()
```

```python
plt.title('Random chart')
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'orange', linestyle=':', alpha=0.9, marker="*", markerfacecolor='y', markersize=20, markeredgecolor='blue',
linewidth = 5)
plt.show()
```

```python
x = np.arange(1, 11)
y = np.sin(x)
z = np.cos(x)
print(x, y, z, sep='\n')
```

```python
plt.plot(x, y, label='sinx', color='green')
plt.plot(x, z, label='cosx', color='black')
plt.title('sinx vs cosx graph')
plt.xlabel('data points')
plt.ylabel('associate values')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,3))
plt.plot(x, y, label='sinx', color='green')
plt.plot(x, z, label='cosx', color='black')
plt.title('sinx vs cosx graph')
plt.xlabel('data points')
plt.ylabel('associate values')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,3))

plt.subplot(1, 2, 1)
plt.plot(x, y, label='sinx', color='green')
plt.plot(x, z, label='cosx', color='black')
plt.title('sinx vs cosx graph')
plt.xlabel('data points')
plt.ylabel('associate values')
plt.legend()

plt.subplot(1,2,2)
plt.plot(x, y, label='sinx', color='green')
plt.plot(x, z, label='cosx', color='black')
plt.title('sinx vs cosx graph')
plt.xlabel('data points')
plt.ylabel('associate values2')
plt.legend()

plt.show()                                                       
```

```python
x = np.arange(1, 60)
y = np.sin(x)
z = np.cos(x)

# print(x, y, z, sep='\n')

plt.figure(figsize=(16,6))
plt.grid()
plt.subplot(3, 3, 1)
plt.grid()
plt.plot(x, y, label='sinx', color='green')
plt.plot(x, z, label='cosx', color='black')
plt.title('sinx vs cosx graph1')
plt.xlabel('data points')
plt.ylabel('associate values')
plt.legend()

plt.subplot(3,3,5)
plt.plot(x, y, label='sinx', color='orange')
plt.plot(x, z, label='cosx', color='brown')
plt.title('sinx vs cosx graph2')
plt.xlabel('data points')
plt.ylabel('associate values2')
plt.grid()
plt.legend()

plt.subplot(3,3,9)
plt.plot(x, y, label='sinx', color='red')
plt.plot(x, z, label='cosx', color='blue')
plt.title('sinx vs cosx graph3')
plt.xlabel('data points')
plt.ylabel('associate values3')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()                                                       
```
