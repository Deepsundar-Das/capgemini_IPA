#### **Matplotlib**

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
x = np.random.randint(1,50,10)
y = np.random.randint(1, 50, 10)
print(x, y, sep='\n')
```

```python
plt.plot(x, y, color = 'brown', linestyle='dotted')
plt.show()
```

```python
plt.title('Random chart')
plt.plot(x, y, color = 'brown', linestyle='dotted')
plt.show()
```

```python
plt.title('Random chart')
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.plot(x, y, color = 'brown', linestyle='dotted')
plt.show()
```

```python
plt.title('Random chart')
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'brown', linestyle='dotted')
plt.show()
```

```python
plt.title('Random chart')
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'orange', linestyle=':')
plt.show()
```

```python
plt.title('Random chart')
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'orange', linestyle=':', alpha=0.9)
plt.show()
```

```python
plt.title('Random chart')
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'orange', linestyle=':', alpha=0.9, marker="*")
plt.show()
```

```python
plt.title('Random chart')
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'orange', linestyle=':', alpha=0.9, marker="*", markerfacecolor='y')
plt.show()
```

```python
plt.title('Random chart')
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'orange', linestyle=':', alpha=0.9, marker="*", markerfacecolor='y', markersize=20)
plt.show()
```

```python
plt.title('Random chart')
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'orange', linestyle=':', alpha=0.9, marker="*", markerfacecolor='y', markersize=20, markeredgecolor='blue')
plt.show()
```

```python
plt.title('Random chart')
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'orange', linestyle=':', alpha=0.9, marker="*", markerfacecolor='y', markersize=20, markeredgecolor='blue',
linewidth = 5)
plt.show()
```

```python
x = np.arange(1, 11)
y = np.sin(x)
z = np.cos(x)
print(x, y, z, sep='\n')
```

```python
plt.plot(x, y, label='sinx', color='green')
plt.plot(x, z, label='cosx', color='black')
plt.title('sinx vs cosx graph')
plt.xlabel('data points')
plt.ylabel('associate values')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,3))
plt.plot(x, y, label='sinx', color='green')
plt.plot(x, z, label='cosx', color='black')
plt.title('sinx vs cosx graph')
plt.xlabel('data points')
plt.ylabel('associate values')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,3))

plt.subplot(1, 2, 1)
plt.plot(x, y, label='sinx', color='green')
plt.plot(x, z, label='cosx', color='black')
plt.title('sinx vs cosx graph')
plt.xlabel('data points')
plt.ylabel('associate values')
plt.legend()

plt.subplot(1,2,2)
plt.plot(x, y, label='sinx', color='green')
plt.plot(x, z, label='cosx', color='black')
plt.title('sinx vs cosx graph')
plt.xlabel('data points')
plt.ylabel('associate values2')
plt.legend()

plt.show()                                                       
```

```python
x = np.arange(1, 60)
y = np.sin(x)
z = np.cos(x)

# print(x, y, z, sep='\n')

plt.figure(figsize=(16,6))
plt.grid()
plt.subplot(3, 3, 1)
plt.grid()
plt.plot(x, y, label='sinx', color='green')
plt.plot(x, z, label='cosx', color='black')
plt.title('sinx vs cosx graph1')
plt.xlabel('data points')
plt.ylabel('associate values')
plt.legend()

plt.subplot(3,3,5)
plt.plot(x, y, label='sinx', color='orange')
plt.plot(x, z, label='cosx', color='brown')
plt.title('sinx vs cosx graph2')
plt.xlabel('data points')
plt.ylabel('associate values2')
plt.grid()
plt.legend()

plt.subplot(3,3,9)
plt.plot(x, y, label='sinx', color='red')
plt.plot(x, z, label='cosx', color='blue')
plt.title('sinx vs cosx graph3')
plt.xlabel('data points')
plt.ylabel('associate values3')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()                                                       
```

#### **Bar Graph**

```python
company = [12,15,20,22]
revenue = [11,34,45,54]
plt.hist(company)
plt.title("Bar plot")
plt.xlabel("Company")
plt.ylabel("revenew")
plt.show()
```

#### **Pie Graph**

```python
plt.pie(revenue)
plt.pie(revenue, autopct='%.3f%%')
plt.title('pie')
plt.show()
```

```python
company = ['A','B','C','D']
revenue = [11,54,76,23]
plt.plot(company, revenue)
plt.xlabel("company")
plt.ylabel('revene')
plt.title('line plot')
```

```python
plt.boxplot(revenue)
plt.title("box plot")
plt.show()
```

```python
ls = [23,46,26, 33]
np.mean(ls)
plt.boxplot(ls)
plt.title("box plot")
plt.show()
```

```python
dff = pd.read_csv("tips.csv")
dff.head()
```

```python
plt.scatter('total_bill', "tip", data = dff, color = "green")
```

```python
dff.day.value_counts()
```

```python
plt.bar(x=dff.day.value_counts().index, height=dff.day.value_counts().values)
```

```python
no1 = dff['total_bill'][:31]
no2 = dff['tip'][51:78]

plt.figure(figsize=(10, 8))

# Subplot 1
plt.subplot(3, 3, 1)
plt.plot(no1)
plt.title("Total Bill Line")

# Subplot 2
plt.subplot(3, 3, 3)
plt.plot(no2)
plt.title("Tip Line")

# Subplot 3
plt.subplot(3, 3, 5)
min_len = min(len(no1), len(no2))
plt.scatter(no1[:min_len], no2[:min_len])
plt.title("Scatter: Total Bill vs Tip")

# Subplot 4
plt.subplot(3, 3, 7)
plt.hist(no1)
plt.title("Total Bill Histogram")

# Subplot 5
plt.subplot(3, 3, 9)
plt.hist(no2)
plt.title("Tip Histogram")

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('sugarcane.csv')
df
```

```python
df.drop(columns='Unnamed: 0', inplace= True)
df
```

```python
df.info()
```

```python
df['Production (Tons)']=df['Production (Tons)'].str.replace('.','')
df.head()
```

```python
df['Production per Person (Kg)']=df['Production per Person (Kg)'].str.replace('.','(:')
df.head()
```

```python
df['Acreage (Hectare)']=df['Acreage (Hectare)'].str.replace('.','😀')
df.head()
```

```python
df['Yield (Kg / Hectare)']=df['Yield (Kg / Hectare)'].str.replace('.','🫠')
df.head()
```

```python
df.rename(
    columns={
        'Production (Tons)' :'🤐',
        'Production per Person (Kg)' : '🥰',
         'Acreage (Hectare)' : '🤗',
        'Yield (Kg / Hectare)' : '🫣'
    },
    inplace = True
)
df.head()
```

```python
df.isnull().sum()
```

```python
df[df.🤗.isnull()]
df
```

```python
df.dropna(inplace = True)
df
```

```python
df.reset_index(drop=True, inplace = True)
df
```

```python
df.columns
```

```python
for i in df.columns[2:]:
    df[i] = df[i].astype('float')
```
