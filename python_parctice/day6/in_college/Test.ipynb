```python
from abc import ABC, abstractmethod
class BankAccount(ABC):
    def __init__(self, customer_name:str, dob:str):
        self.customer_name = customer_name
        self.dob = dob
        
    @abstractmethod
    def deposit(self,amount):
        pass
    @abstractmethod
    def withdraw(self, amount):
        pass
        

```

```python
class SavingAccount(BankAccount):
    def __init__(self, customer_name:str, dob:str, initial_amount:int):
        super().__init__(customer_name, dob)
        self.initial_amount = initial_amount
        
    def deposit(self, amount):
        if amount <= 0:
            print("negative or zero amount cannot be deposited")
        else:
            self.initial_amount += amount
            print(f"your current balance is {self.initial_amount}")

    def withdraw(self, amount):
        if amount <= 0:
            print(f"negative or zero amount cannot be withdrawed, your current balance is {self.initial_amount}")
        elif amount > self.initial_amount:
            print(f"your current balance is {self.initial_amount}, you cannot transfer {amount}")
        else:
            self.initial_amount -= amount
            print(f"your current balance is {self.initial_amount}")
```

```python
saobj1 = SavingAccount("deep",'02-07-2002',1000)
saobj1.deposit(200)
saobj1.withdraw(500)
saobj1.deposit(0)
```

```python
class CurrentAccount(BankAccount):
    def __init__(self, customer_name:str, dob:str, current_amount:int):
        super().__init__(customer_name, dob)
        self.current_amount = current_amount
        self.overdraft_limit = 1000
        
    def deposit(self, amount):
        if amount <= 0:
            print("negative or zero amount cannot be deposited")
        else:
            self.current_amount += amount
            print(f"your current balance is {self.current_amount}")

    def withdraw(self, amount):
        if amount <= 0:
            print(f"negative or zero amount cannot be withdrawed, current balance is {self.current_amount}")
        elif amount > (self.current_amount+self.overdraft_limit):
            print(f"your current balance is {self.current_amount} and your overdraft limit is {self.overdraft_limit}, you cannot transfer {amount}")
        else:
            if(amount > self.current_amount):
                amount -= self.current_amount
                self.current_amount = 0
                self.overdraft_limit -= amount
            else:
                self.current_amount -= amount
            print(f"your current balance is {self.current_amount}")
```

```python
cur1 = CurrentAccount('deep','09-08-2002',1000)
cur1.deposit(2000)
cur1.withdraw(00)
```
