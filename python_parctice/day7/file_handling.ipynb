##### **count number of words in a file**

```python
with open('test.txt') as f:
    words = f.read().split(',')
    print(len(words))
```

```python
with open('test.txt','r+') as f:
    set1 = set()
    words = f.read().split()
    setwords = set(words)
    for i in words:
        set1.add(i)

    # print(type(words))
    # set1.add(tuple(words))
    print(set1)
```

##### **create a dictionary of word count**

```python
from collections import defaultdict
with open('test.txt', 'r') as f:
    words = f.read().split()
    dictionary = defaultdict(int)
    for i in words:
        dictionary[i] +=1
    print(dictionary)
```

#### **csv file**

```python
from csv import writer
import csv 
with open ('data.csv','w', newline="") as file:
    writer = csv.writer(file)
    writer.writerow(['name', 'age'])
    writer.writerow(['deep',23])
```

```python
rows = (
["name","age"],
["deep",23],
["arka",22],
["subh",23]
)
with open('data.csv','w', newline="\n") as file:
    writer = csv.writer(file)
    writer.writerow(rows)
```

```python
with open('data1.csv','w',newline="") as file:
    fieldNames=["name","age"]
    writer = csv.DictWriter(file,fieldnames=fieldNames)
    writer.writeheader()
    writer.writerow({"name":"deep","age":23})
```

```python
with open ('data.csv','r') as file:
    reader = csv.DictReader(file)
    print(reader)
    for row in reader:
        print(row)
```
